```python
import plotly.express as px
df1 = px.data.iris()
df1
fig1 = px.line(df1,
               x='sepal_length',
               y='petal_length',
               title='Sepal Length vs Petal Length',
               markers=True)
fig1.show()
fig2=px.box(df1,
            x='species',
            y='sepal_width',
            title = 'Sepal Length Distribution by Species')
fig2.show()
```
